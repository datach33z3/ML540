```python
import numpy as np
import matplotlib.pyplot as plt
import skfuzzy as fuzz
from skfuzzy import control as ctrl
```

```python
## Number 2 P1
q = np.linspace(-10,10,60) 
s = np.linspace(-10,10,60) 
```

```python
y = (np.sin(q) * np.sin(s))/(q * s)
plt.plot(q,y)
plt.show()
```

```python
#Setting Antecendent objects 
q = ctrl.Antecedent(np.arange(-10, 10, 1), 'q')
s = ctrl.Antecedent(np.arange(-10, 10, 1), 's')
y = ctrl.Consequent(y, 'y')
```

```python
#Now we set the membership functions for three memberships. 
q.automf(3)
s.automf(3)
```
